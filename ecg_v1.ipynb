```
#making the NN as per paper description

import keras
from keras import regularizers
from keras.layers import Dense,Activation,Flatten
from keras.models import Sequential
from keras.layers.convolutional import Conv1D,MaxPooling1D
from keras.regularizers import l2
```

```
##working with ZS
data = np.array(data_zs)
features = data[:, :1]
dataset = np.delete(data, 0, axis=1)
features = np.where(features == 'Z', 0, features)
features = np.where(features == 'S', 1, features)
print(features)
```

```
X_train, X_test, Y_train, Y_test = train_test_split(dataset, features, test_size=0.3)
X_train = X_train.reshape(X_train.shape[0], 4097, 1)
X_test = X_test.reshape(X_test.shape[0], 4097, 1)
```

```
num_classes = 2
model = Sequential()
model.add(Conv1D(filters=4, kernel_size=6, strides=1, input_shape=(4097, 1)))
model.add(MaxPooling1D(pool_size=2, strides=2))

model.add(Conv1D(filters=4, kernel_size=5, strides=1))
model.add(MaxPooling1D(pool_size=2, strides=2))

model.add(Conv1D(filters=4, kernel_size=4, strides=1))
model.add(MaxPooling1D(pool_size=2, strides=2))

model.add(Conv1D(filters=10, kernel_size=4, strides=1))
model.add(MaxPooling1D(pool_size=2, strides=2))

model.add(Conv1D(filters=10, kernel_size=4, strides=1))
model.add(MaxPooling1D(pool_size=2, strides=2))

model.add(Flatten())

from keras.layers import LeakyReLU

model.add(Dense(50, activation='linear', kernel_regularizer=l2(0.7)))
model.add(LeakyReLU(alpha=0.1))
model.add(Dense(1, activation='softmax'))

model.add(Dense(20, activation='linear', kernel_regularizer=l2(0.7)))
model.add(LeakyReLU(alpha=0.1))
model.add(Dense(1, activation='softmax'))

model.add(Dense(2, activation='linear', kernel_regularizer=l2(0.7)))
model.add(LeakyReLU(alpha=0.1))
model.add(Dense(1, activation='softmax'))

from keras.optimizers import SGD

learning_rate = 0.0001

momentum = 0.3

sgd = SGD(lr=learning_rate, momentum=momentum, nesterov=False)
model.compile(loss='binary_crossentropy', optimizer=sgd, metrics=['accuracy'])
```

```
model.fit(X_train, Y_train, validation_data=(X_test, Y_test), epochs=150, batch_size=3)
score = model.evaluate(X_test, Y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```
import keras.backend as K


def get_activations(model, model_inputs, print_shape_only=False, layer_name=None):
    print('----- activations -----')
    activations = []
    inp = model.input

    model_multi_inputs_cond = True
    if not isinstance(inp, list):
        # only one input! let's wrap it in a list.
        inp = [inp]
        model_multi_inputs_cond = False

    outputs = [layer.output for layer in model.layers if
               layer.name == layer_name or layer_name is None]  # all layer outputs

    funcs = [K.function(inp + [K.learning_phase()], [out]) for out in outputs]  # evaluation functions

    if model_multi_inputs_cond:
        list_inputs = []
        list_inputs.extend(model_inputs)
        list_inputs.append(0.)
    else:
        list_inputs = [model_inputs, 0.]

    # Learning phase. 0 = Test mode (no dropout or batch normalization)
    # layer_outputs = [func([model_inputs, 0.])[0] for func in funcs]
    layer_outputs = [func(list_inputs)[0] for func in funcs]
    for layer_activations in layer_outputs:
        activations.append(layer_activations)
        if print_shape_only:
            print(layer_activations.shape)
        else:
            print(layer_activations)
    return activations


print('0')
print(model.layers[0].output)
print('1')
print(model.layers[1].output)
print('2')
print(model.layers[2].output)
print('3')
print(model.layers[3].output)
print('4')
print(model.layers[4].output)
print('5')
print(model.layers[5].output)
print('6')
print(model.layers[6].output)
print('7')
print(model.layers[7].output)
print('8')
print(model.layers[8].output)
print('9')
print(model.layers[9].output)
print('10')
print(model.layers[10].output)
print('11')
print(model.layers[11].output)
print('12')
print(model.layers[12].output)
print('13')
print(model.layers[13].output)
print('14')
print(model.layers[14].output)
print('15')
print(model.layers[15].output)
print('16')
print(model.layers[16].output)
```
